# 가변 시퀀스 메서드(리스트 전용)
L.append(x) : append 자체는 반환값이 없다. print(list.append())하면 none임
L.extend(iterable) : list.extend([1,2,3]) 하면 반복가능한 객체 하나하나를 추가(숫자는 반복불가해서 추가못함). 
L.insert(i,x) : i인덱스에 x를 넣는다. append는 끝에 넣는거면 이건 특정위치에 넣을 수 있다.
L.remove(x) : 리스트에서 첫번쨰로 일치하는 항목을 삭제
L.pop() 또는 L.pop(i) : 괄호안 값을 제거(공백시 끝값) 하고 그 값을 반환.
L.clear() : 리스트의 모든 항목을 제거(빈리스트로)

# 리스트 정렬 메서드
L.reverse() : 리스트의 순서를 역순으로 변경(정렬x). 이것 자체의 반환갑은없음.
L.sort() : 원본 리스트를 오름차순 정렬 / 내림차순: reverse=True
- sorted() 는 내장함수이고, L.sort()는 리스트의 메서드임(리스트전용). 그리고 얘는 반환값있지만,, sort는 없음.

# 복사
- 가변/불변 객체의 개념
  가변객체: 생성후 내용을 변경할 수 있는 객체
  불변객체: 생성 후 내용을 변경할 수 없는 객체

변수 할당의 의미: 사실 변수에는 메모리 주소가 들어있다. 메모리주소를 외울 수 없기 때문에 이름표를 붙인다. 

변수가 가지고 있는건 값이 아니라 주소이다! 
id() 함수를 사용한 메모리 주소 확인

## 얕은복사
객체의 최상위 요소만 새로운 메모리에 복사하는 방법. 내부에 중첩된 객체가 있다면 그 객체의 참조만 복사됨.
1. 리스트 슬라이싱
2. copy() 메서드
3. list() 함수
그니까 중첩된 객체가 하나라도 있으면 그 중첩객체는 서로 같은 메모리 주소를 바라봄. 

## 깊은복사
객체의 모든 수준의 요소를 새로운 메모리에 복사하는 방법. 중첩된 객체까지 모두 새로운 객체로 생성됨.
copy 모듈에서 제공하는 deepcopy() 함수를 사용
import copy
a = copy.deepcopy(b)
