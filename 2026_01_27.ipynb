# 비시퀀스 데이터 구조
*딕셔너리*, 세트

# 딕셔너리
.get(key[,default])
키 연결된 값을 반환하거나 키가 없으면 None 혹은 기본값을 반환
dict['name'] 과 같이 값을 반환하는건데, 키가 없어도 값을 지정할 수 있다는 것.
.keys() : map,range,zip처럼 덩어리로 일단 저장함.
프린트하면 덩어리형태로 나오지만, for문에서 리스트처럼 in뒤에 활용가능
.values()
.items()
.pop(key[,default]) : 키를 제거하고 연결됐던 값을 반환. (없으면 에러나 default 반환)
.clear()
.setdefault(key[,default]) : 키와 연결된 값을 반환, 키가 없다면 뒤에 값을 추가해줌. 그리고 그 값을 반환.
.update([other]) : other는 키-값 쌍을 의미. 딕셔너리 자체를 넣을 수도 있고 키=값 형태로 추가도 가능.

# 세트
딕셔너리와 같이 비시퀀스 데이터인데, 내부적 해시테이블을 사용하여 데이터를 저장함.
.add(x) : 순서이상
.update(iterable) : 리스트같은 형태의 요소들 추가가능. 순서이상.
.clear()
.remove(x)
.pop() : 임의의 요소를 제거하고 반환
.discard(x) : 얘도 remove랑 같은데 없는거 입력하면 에러안뜨는점.
## 세트의 집합 메서드
set1.difference(set2)
set1.intersection(set2)
set1.issubset(set2)
set1.issuperset(set2)
set1.union(set2)

# 참고
- 딕셔너리의 확장:defaultdict
  파이썬 내장모듈 collections에서 제공하는 딕셔너리의 확장판
  defaultdict(자료형) 
  조회만해도 키가 생성되기 땜에, 조회만하고싶으면 get.
- 파이썬 문법 규격
  [] : 선택적 요소
  {} : 0번 이상 반복
  () : 그룹화
- 해시 테이블 : 키와 값을 짝지어 저장하는 자료구조
  임의의 크기를 가진 데이터(키)를 고정된 크기의 고유의 값으로 변환하는것
  해시함수: 임의 길이 데이터를 입력 받아 고정길이(정수)로 변환해주는 함수. 이 정수가 바로 해시값.
  해시 값을 인덱스로 삼아 데이터를 저장.검색
  파이썬의 set은 순서가 없고 pop할때 어떤요ㅕ소가 반환될지 모름
  dict는 삽입순서는 보장되지만, 내부구현은 여전히 해시테이블
