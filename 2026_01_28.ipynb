## 절차지향/객체지향
절차지향: 어떤 순서로 처리할까?
객체지향: 어떤 객체가 이 문제를 해결할까?/이 객체는 어떤 속성과 기능을 가질까?
객체지향은 수동적인 데이터가 능동적인 객체로 변화한것
객체 지향에서는 데이터와 해당 데이터를 처리하는 메서드가 하나의 객체로 통합되어 스스로 기능을 수행하는 능동적 존재가됨
실제 세계의 모델링 방식과 더 유사한 프로그래밍을 가능하게 함
절차 지향과 객체 지향은 대조되는 개념이 아니다: 객체 지향은 기존 절차 지향을 기반으로 두고 보완하기 위해 객체라느 ㄴ개념을 도입해 상속,코드 재사용성, 유지보수성 등의 이점을 가지느 ㄴ패러다임.

# 객체와 클래스
객체(Object): 실제 존재하는 사물을 추상화한 것 "속성(변수(데이터))"과 "동작(메서드)"을 가짐.
"강아지"라는 객체: 속성:이름,종,나이 행동:뛰기,먹기
클래스(Class): 객체를 만들기 위한 설계도. 데이터와 기능을 함께 묶는 방법을 제공. 파이썬에서 타입을 표현하는 방법.
객체: 속성(변수) / 동작(메서드)
가수(클래스) > 객체(아이유,BTS...)
각각의 객체는 고유의 특성을 가짐.

# 클래스 기초
클래스(Class): 클래스는 관련된 데이터(속성)와 기능(메서드)을 하나의 "묶음"으로 정의하는 설계도(blueprint)
1. 클래스 == 자동차설계도
2. 속성 == 자동차의 데이터 (색상,최고속도,좌석수,배기량 등)
3. 매서드 == 자동차의 기능 ( 가속하기(),정지하기())
4. 인스턴스 == 실제 자동차
클래스 정의: class키워드, 클래스 이름은 파스칼(Pascal Case) 케이스 방식으로 작성(원레는 snake_case). 각 단어의 앞문자만 대문자로.
함수면 my_class 클래스는 MyClass.
클래스 안에 언더바 붙은 메서드들은 매직메서드 라고 함. 이거는 개발자가 호출에 관여하지 않음. 얘네들은 호출시점을 파이썬이 결정.

# 인스턴스
인스턴스(Instance) : 클래스를 통해 생성된 객체
인스턴스는 클래스는 사용해 실제로 만들어진 객체입니다. 같은 클래스로 여러 인스턴스를 만들 수 있으며, 각 인스턴스는 클래스 구조를 따라 동작하지만 서로 독립된 데이터를 가질 수 있습니다.
## 인스턴스 예시
p1 = Person("Alice",25) #p1은 인스턴스이고, Person이라는 클래스
p1.introduce() : Person클래스의 메서드를 이용.

가수(클래스) > 아이유는 객체다(O)/아이유는 인스턴스다(X)/아이유는 가수의 인스턴스다(O)

클래스와 인스턴스: 클래스를 정의한다는 것은 공통된 특성과 기능을 가진 틀을 만드는 것. 실제로 활동하는 개별 객체들은 이 틀에서 생성된 인스턴스. 공통된 특성과 기능을 가진 틀을 만드는 것은 곧 새로운 타입을 만드느 행위.

# 메서드
클래스 내부에 정의된 함수로, 해당 객체가 어떻게 동작할지를 정의
메서드는 단독호출이 불가능. / 객체.메서드 / 정확하게는 인스턴스.메서드

## 메서드 종류
1. 인스턴스 메서드
   인스턴스의 상태를 조작하거나 동작을 수행합니다.
   계산기 객체가 있고 그 안에 더하기 기능이 있다면, 사용자가 맏
   인스턴스마다 독립적으로 행동할 수 있게 만드는 것이 인스턴스 메서드.
   - 인스턴스 메서드 구조(반드시 첫번쨰 인자로 인스턴스 자신(self)을 받음)
    'hello'.upper() > str.upper('hello') 로 동작
    def adafd(self) ~
    

2. 클래스 메서드
   클래스 변수를 조작하거나 클래스 레벨의 동작을 수행합니다.
   모든 학생이 다니는 '학교 이름'을 설정하는 기능이 있다고 해봅시다. 이럴 때는 학생 각각이 아니라, 클래스 전체에 적용되는 메서드가 필요합니다. 
   @calssmethod
   def class_method(cls, arg1, ...):
   pass

3. 스태틱 메서드
   클래스, 인스턴스와 상관없이 독립적으로 동작하는 메서드.

# 메서드 정리
- 인스턴스 메서드 : 인스턴스의 상태를 변경하거나, 해당 인스턴스의 특정 동작을 수행
- 클래스 메서드: 인스턴스의 상태에 의존하지 않는 기능을 정의, 클래스 변수를 조작하거나 클래스 레벨의 동작을 수행
- 스태틱 메서드: 클래스 및 인스턴스와 관련없는 일반동작을 수행

클래스가 사용해야 할 것: 클래스메서드, 스태틱 메서드
인스턴스가 사용해야 할 것: 인스턴스 메서드
